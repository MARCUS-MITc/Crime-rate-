```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("CRIME_REVIEW_FOR_THE_MONTH_OF_APRIL_2022.csv")
```

```python
df.head()
df.columns
```

```python
df.shape
```

```python
plt.figure(figsize=(10, 8))
df.plot(x='During the current year upto the end of month under review', y='During the corresponding month of previous year',kind='scatter')
```

As we can see there is a linear relation between these axes we can say that the cases under review this year either increased or remain same from the previous year in same month which means courts are not efficient in solving the judicial problems. 

```python
df.plot(x='During the previous month', y='During the current month',kind='scatter')
```

As we can see there is a linear relation between these axes we can say that the cases during this month either increased or remain same from the previous month which means courts are not efficient in solving the judicial problems or the no of courts in india not enough to tackle the cases as the cases are increasing linearly where as no of courts are remained same.

```python
axes = df.plot.line(subplots=True)
for ax in axes:
    fig = ax.get_figure()
    fig.set_figheight(8)
    fig.set_figwidth(12)
plt.show()
```

as we can see all these line plots have same relation between all the data whether its previous year or present year and previous month or present month.
